```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Function to compute the harmonic number
harmonic_number = lambda N: np.sum(1 / np.arange(1, N + 1))

# Function to compute variance
harmonic_number_squared = lambda N: np.sum(1 / np.arange(1, N + 1)**2)

# Calculate expectation and variance for each N
def expected_stickers(N):
    return N * harmonic_number(N)

def variance_stickers(N):
    # return N**2 * harmonic_number_squared(N)
    return N**2 * harmonic_number_squared(N) - N * harmonic_number(N)

# Define a range of unique stickers to analyze
N_values = range(1, 51, 1)  # from 1 to 50, step 1

# Create a DataFrame with the expectations and variance
data = {
    'N (Unique Stickers)': N_values,
    'Expected Total Stickers Needed': [expected_stickers(N) for N in N_values],
    'Variance': [variance_stickers(N) for N in N_values],
    'Standard Deviation': [np.sqrt(variance_stickers(N)) for N in N_values]
}

expectation_df = pd.DataFrame(data)

# Print the datafram to look at specific values
print(expectation_df)

# Plotting
plt.figure(figsize=(12, 6))

# Plot Expected Total Stickers
plt.subplot(1, 3, 1)
plt.plot(expectation_df['N (Unique Stickers)'], expectation_df['Expected Total Stickers Needed'], marker='o')
plt.title('Expected Total Stickers Needed by Album Size')
plt.xlabel('N (Unique Stickers)')
plt.ylabel('Expected Stickers')
plt.grid(True)

# Plot Variance
plt.subplot(1, 3, 2)
plt.plot(expectation_df['N (Unique Stickers)'], expectation_df['Variance'], marker='o', color='green')
plt.title('Variance by Album Size')
plt.xlabel('N (Unique Stickers)')
plt.ylabel('Variance')
plt.grid(True)

plt.tight_layout()
plt.show()

# Plot Standard Deviation
plt.subplot(1, 3, 3)
plt.plot(expectation_df['N (Unique Stickers)'], expectation_df['Standard Deviation'], marker='o', color='red')
plt.title('Standard Deviation by Album Size')
plt.xlabel('N (Unique Stickers)')
plt.ylabel('Standard Deviation')
plt.grid(True)

plt.tight_layout()
plt.show()
```
